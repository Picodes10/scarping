```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

# Target URL
url = "https://www.noon.com/uae-en/sports-and-outdoors/exercise-and-fitness/yoga-16328/"

# Headers to mimic a browser
headers = {"User-Agent": "Mozilla/5.0"}

# Retry strategy
retry_strategy = Retry(
    total=3,
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=["HEAD", "GET", "OPTIONS"]
)
adapter = HTTPAdapter(max_retries=retry_strategy)
http = requests.Session()
http.mount("https://", adapter)
http.mount("http://", adapter)

# Send GET request
response = http.get(url, headers=headers, timeout=10)
soup = BeautifulSoup(response.content, "html.parser")

# Extract product details
products = []
for item in soup.find_all("div", class_="product-container-class"):  # Replace with correct class
    try:
        name = item.find("h3", class_="product-title-class").text.strip()  # Replace with correct class
        price = item.find("span", class_="price-class").text.strip()  # Replace with correct class
        brand = item.find("div", class_="brand-class").text.strip()  # Replace with correct class
        seller = item.find("div", class_="seller-class").text.strip()  # Replace with correct class
        products.append({"Name": name, "Price": price, "Brand": brand, "Seller": seller})
    except AttributeError:
        continue

# Save data to a CSV file
df = pd.DataFrame(products)
df.to_csv("data/products.csv", index=False)
```
